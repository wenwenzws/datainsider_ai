```python
import pandas as pd
import plotly as pl
import plotly.graph_objs as go
import report
import plotly.express as px
```

```python
uploaded_file='DynamicsExport_638582166632440668.xlsx'
```

```python
df = pd.read_excel(uploaded_file)
list_of_columns = df.columns.tolist()
df['Date'] = pd.to_datetime(df['Date'])
# Add a column for the weekday name
df['Weekday'] = df['Date'].dt.day_name()
df['Sales dollars before tax'] = abs(df['Amount before tax'])
df['Items_count'] = 1
```

```python
df
```

```python
target='Sales dollars before tax'
time = 'Date'
```

```python
daily_df = df.groupby([pd.Grouper(key='Date', freq='ME'),'Operator ID'])[target].sum().reset_index()
fig = px.line(daily_df, x="Date", y=target, color='Operator ID')
fig
```
